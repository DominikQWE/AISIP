```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Завантаження (файл вже перевірено)
file_name = 'satir.jpg' 
img = cv2.imread(file_name, cv2.IMREAD_GRAYSCALE)

# 2. Математичні операції
# Негатив: міняємо місцями 0 та 255
img_neg = 255 - img

# Лінійне контрастування: розтягуємо гістограму на весь діапазон [0, 255]
img_contrast = cv2.normalize(img, None, 0, 255, cv2.NORM_MINMAX)

# Побудова гістограми оригіналу
hist = cv2.calcHist([img], [0], None, [256], [0, 256])

# 3. Вивід результатів для звіту
plt.figure(figsize=(15, 10))

# Оригінал
plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('1. Оригінальне зображення')
plt.axis('off')

# Негатив
plt.subplot(2, 2, 2)
plt.imshow(img_neg, cmap='gray')
plt.title('2. Негатив (Інверсія)')
plt.axis('off')

# Контрастування
plt.subplot(2, 2, 3)
plt.imshow(img_contrast, cmap='gray')
plt.title('3. Лінійне контрастування')
plt.axis('off')

# Гістограма
plt.subplot(2, 2, 4)
plt.plot(hist, color='black')
plt.fill_between(range(256), hist.flatten(), color='gray', alpha=0.3)
plt.title('4. Гістограма яскравості')
plt.xlabel('Тон (0-255)')
plt.ylabel('Кількість пікселів')

plt.tight_layout()
plt.show()
```
